```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import missingno as msno
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
y = train_data["Survived"]
train_data.pop('Survived')
```

```python
def InfoPrepros(DataFrame):
    X = DataFrame.loc[0:,["Pclass", "Sex", "SibSp", "Parch","Age"]]
    X.fillna(X["Age"].median(), inplace = True) 
    X = pd.get_dummies(X, columns = ["Pclass", "Sex", "SibSp", "Parch"], drop_first = True)
    return X


data = pd.concat([train_data, test_data], ignore_index=True, sort = False, axis = 0)
data = InfoPrepros(data)
X_train = data.loc[0:len(train_data)-1]
X_test = data.loc[len(train_data):(len(train_data)+len(test_data)-1)]
```

```python
from sklearn.ensemble import RandomForestClassifier



model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
